# 확률 분포

이산확률분포 : 이산확률변수에 의해 만들어진 함수 분포<br>
연속확률분포 : 연속확률변수에 의해 만들어진 함수 분포<br>

어떤 이산확률분포도든지 무한 번 반복하다 보면 대부분 어떤 연속확률분포 중의 한 형태로 수렴하게 된다.
- 비교적 적은 수의 시행만으로도 무한 번 시행한 것과 같은 확률의 분포를 추측해낼 수 있다. 

기댓값(평균) : 분포의 중심(분포에서 가장 많은 양을 가진 값).<br>
편차 : 데이터와 평균값의 차이.<br>
분산 : 멀어진 거리를 제곱한 것의 평균., 확률변수가 기댓값으로부터 얼마나 떨어진 곳에 분포하는지를 가늠하는 숫자<br>
표준편차 : 
- 분포가 흩어진 정도(평균을 중심으로 데이터들이 얼마나 멀리 떨어져 있는지)<br>
- 편차들을 제곱해서 평균을 취한 다음에 루트를 씌운 것 -> 평균적인 편차 

분산을 쓰는 이유
- 편차의 __"개념"__ 은 음의 값을 가질 수 없기에 절대값을 이용하여 퍼진 정도를 측정한다.<br>
- 그러나 절대값을 사용하는 경우에 통계량의 수학적 조작(미분 등) 불가능한 경우가 생긴다.
